# Break


```python
for c in range(10):
    if c == 5:
        break  # CANCELA o loop
    print(c)

print('fim')
```

# Continue


```python
for c in range(10):
    if c == 5:
        continue  # CANCELA a iteração
    print(c)

print('fim')
```

```python
# Quero saber se tem um valor dentro do range que vale 5
igual = False

for c in range(100000000000):
    if c == 5:
        igual = True
        break

print(f'{igual=}')
```

```python
x = 0
while x < 100:  # Não cria a variável
    print(f'{x=}')
    if x % 5 == 0:  # 100%5==0 True
        print(x)
    x += 1

    # x=100
```

```python
soma = quantidade = 0

# quantidade+=1
# soma+=num
num = int(input("Insira um numero: "))
# quantidade = 1
# soma = num5
while num >= 0:
    quantidade += 1
    soma += num
    print(f'{soma=}, {quantidade=}, média = {soma/quantidade}')
    num = int(input("Insira um numero: "))
```

# Fatorial

```python
x = int(input("Digite um numero: "))
while x<=0 or x>=20:
    x = int(input("Digite um numero valido entre 1-19: "))
    multiplicacao = x
    if x>0 and x<20:
        for x in range(1, x+1, 1):
            multiplicacao *= x
            print(multiplicacao)
```
